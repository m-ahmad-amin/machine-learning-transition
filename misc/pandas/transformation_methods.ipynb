```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name": ["Ali", "Sana", "Saba", "Fahad", "Aryan"],
    "Gender": ["Male", "Female", "Female", "Male", "Male"],
    "Score": [98, 78, 95, 64, 87]
})
```

```python
df
```

```python
df['G'] = df['Gender'].map({'Male': 'M', 'Female': 'F'})
```

```python
df
```

```python
df['Ge'] = df['G'].replace({'M': 'Ma'})
```

```python
df
```

```python
df['Ge'] = df['Ge'].replace({'F': 'Fe'})
```

```python
df
```

```python
df['Grade'] = df['Score'].apply(lambda x: 'Pass' if x > 85 else 'Fail')
```
